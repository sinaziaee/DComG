```python
import numpy as np
import pandas as pd
import os
import pathlib
from dataset_loader import load_dataframes

df_list = [final_dt_df, final_w2v_df, final_nv_df, final_fin_df, final_in_df, final_se_df, all_df] = load_dataframes()
```

```python
base_dir = f'{str(pathlib.Path().resolve())}/datasets'
atc_fname = 'DCDB_ATC codes.csv'
dn_fname = 'dcdb.csv'

path_atc = f'{base_dir}/{atc_fname}'
path_dname = f'{base_dir}/{dn_fname}'

atc_df = pd.read_csv(path_atc)
dn_df = pd.read_csv(path_dname)

drug_list = []
atc_list_1 = []
atc_list_2 = []
atc_list_3 = []
atc_dict = dict()

for i, row in enumerate(atc_df.values):
    drug = row[0]
    atc_codes = str(row[1]).replace('[', '').replace(']', '').replace("'", '').replace(' ', '').split(',')
    drug_list.append(drug)
    atc_list_1.append(atc_codes[0])
    if len(atc_codes) >= 2:
        atc_list_2.append(atc_codes[1])
    else:
        atc_list_2.append('-')
    if len(atc_codes) >= 3:
        atc_list_3.append(atc_codes[2])
    else:
        atc_list_3.append('-')
    atc_dict[drug] = i

a_list1 = []
a_list2 = []
a_list3 = []

for i, row in enumerate(dn_df.values):
    dcc = row[0]
    drug = row[1]
    if dcc in drug_list:
        inx = atc_dict[dcc]
        a_list1.append(atc_list_1[inx])
        a_list2.append(atc_list_2[inx])
        a_list3.append(atc_list_3[inx])
    else:
        a_list1.append('-')
        a_list2.append('-')
        a_list3.append('-')

dn_df['atc1'] = a_list1
dn_df['atc2'] = a_list2
dn_df['atc3'] = a_list3

dn_df.to_csv(f'{base_dir}/drug_atc.csv', index=False)
```

```python
drug_atc_dict = dict()

for row in drugs_with_atc_df.values:
    drug = row[0]
    atc1 = row[1]
    atc2 = row[2]
    drug_atc_dict[drug] = [atc1, atc2]

atc1_list = []
atc2_list = []

for row in dn_df.values:
    drug = row[1]
    atc1 = row[2]
    atc2 = row[3]
    if atc1 == '-':
        if drug in list(drug_atc_dict.keys()):
            atc1_list.append(drug_atc_dict[drug][0])
            atc2_list.append(drug_atc_dict[drug][1])
        else:
            atc1_list.append(atc1)
            atc2_list.append(atc2)
    else:
        atc1_list.append(atc1)
        atc2_list.append(atc2)

dn_df['atc1'] = atc1_list
dn_df['atc2'] = atc2_list
```

```python
code_dict = dict()
code_dict['A'] = 'Alimentary tract and metabolism'
code_dict['B'] = 'Blood and blood forming organs'
code_dict['C'] = 'Cardiovascular system'
code_dict['D'] = 'Dermatologicals'
code_dict['G'] = 'Genito-urinary system and sex hormones'
code_dict['H'] = 'Systemic hormonal preparations, excluding sex hormones and insulins'
code_dict['J'] = 'Antiinfectives for systemic use'
code_dict['L'] = 'Antineoplastic and immunomodulating agents'
code_dict['M'] = 'Musculo-skeletal system'
code_dict['N'] = 'Nervous system'
code_dict['P'] = 'Antiparasitic products, insecticides and repellents'
code_dict['R'] = 'Respiratory system'
code_dict['S'] = 'Sensory organs'
code_dict['V'] = 'Various'
code_dict
```

```python
atc1_first_list = []
atc2_first_list = []
atc1_class_list = []
atc2_class_list = []

for row in dn_df.values:
    dcc = row[0]
    drug = row[1]
    atc1 = row[2]
    atc2 = row[3]
    if atc1 == '-':
        atc1_first_list.append('-')
        atc2_first_list.append('-')
        atc1_class_list.append('-')
        atc2_class_list.append('-')
    else:
        atc1_first_list.append(str(atc1)[0])
        if str(atc1)[0] in list(code_dict.keys()):
            atc1_class_list.append(code_dict[str(atc1)[0]])
        else:
            atc1_class_list.append('-')
        if atc2 != '-':
            atc2_first_list.append(str(atc2)[0])
            if str(atc2)[0] in list(code_dict.keys()):
                atc2_class_list.append(code_dict[str(atc2)[0]])
            else:
                atc2_class_list.append('-')
        else:
            atc2_first_list.append('-')
            atc2_class_list.append('-')

dn_df['atc1_first'] = atc1_first_list
dn_df['atc2_first'] = atc2_first_list
dn_df['atc1_first_class'] = atc1_class_list
dn_df['atc2_first_class'] = atc2_class_list
```
